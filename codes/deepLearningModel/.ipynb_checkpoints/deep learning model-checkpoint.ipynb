```python
# model_0에 배정된 데이터셋을 train, validation, test로 분리
from sklearn.model_selection import train_test_split

x_train_model_0, x_test_model_0, y_train_model_0, y_test_model_0 = train_test_split(x_data_for_model_0, y_data_for_model_0, test_size=0.2, random_state=1)
x_train_model_0, x_val_model_0, y_train_model_0, y_val_model_0 = train_test_split(x_train_model_0, y_train_model_0, test_size=0.25, random_state=1)

# ANN 모델 설계
from tensorflow.keras import models
from tensorflow.keras import layers
from keras.callbacks import EarlyStopping

model_0 = models.Sequential()
model_0.add(layers.Dense(64, activation="relu", input_shape = (41, )))
model_0.add(layers.Dropout(0.6))
model_0.add(layers.Dense(64, activation="relu"))
model_0.add(layers.Dropout(0.6))
model_0.add(layers.Dense(64, activation="relu"))
model_0.add(layers.Dropout(0.6))
model_0.add(layers.Dense(64, activation="relu"))
model_0.add(layers.Dropout(0.6))
model_0.add(layers.Dense(1))

# 조기종료 콜백함수 정의
early_stopping = EarlyStopping(patience = 15) 

# compile 설정
model_0.compile(optimizer="rmsprop", loss = "mse", metrics = ["mae"])

# 해당 model학습
history_model_0 = model_0.fit(x_train_model_0, y_train_model_0, validation_data=(x_val_model_0, y_val_model_0), epochs=300, batch_size= 32, callbacks=[early_stopping])

# 해당 model저장
model_0.save("model_0.h5")

# model_0_result
%matplotlib inline

loss = history_model_0.history["loss"]
mae = history_model_0.history["mae"]

val_loss = history_model_0.history["val_loss"]
val_mae = history_model_0.history["val_mae"]

epoch = list(range(1, len(loss) + 1))

plt.title("Training loss")
plt.plot(epoch, loss, "bo", label = "Training loss")
plt.show()

plt.figure()

plt.title("Validation loss")
plt.plot(epoch, val_loss, "b", label = "Validation loss")

plt.figure()

plt.title("Training mae")
plt.plot(epoch, mae,"bo", label = "Training mae")

plt.figure()

plt.title("Validation mae")
plt.plot(epoch, val_mae, "b", label = "Validation mae")

plt.show()
```
